```python
import requests
from typing import Dict

# REPLACE THE VARIABLES BELOW WITH YOUR DATABRICKS CLOUD URL, TOKEN, AND MAS MODEL SERVING ENDPOINT NAME
DATABRICKS_HOST = '<WORKSPACE_URL>'
DATABRICKS_TOKEN = '<PERSONAL_ACCESS_TOKEN>'
ENDPOINT_NAME = '<MAS_ENDPOINT_NAME>'

# REST API URL for model serving endpoint
url = f"{DATABRICKS_HOST}/serving-endpoints/{ENDPOINT_NAME}/invocations"

# Payload must include the required 'input' field (array) instead of 'messages'
# payload: Dict = {
#     "input": [
#         {"role": "user", "content": "Get me my latest billing status. my name is David Fisher, my email is lbyrd@example.net and my phone number is 001-746-980-4436"}
#     ]
# }
payload: Dict = {
    "input": [
        {"role": "user", "content": "tell me the billing standard policies and when do i make payment?"}
    ]
}
headers = {
    "Authorization": f"Bearer {DATABRICKS_TOKEN}",
    "Content-Type": "application/json"
}

response = requests.post(url, json=payload, headers=headers)
stored_response = ''
if response.status_code == 200:
    result = response.json()
    stored_response = result['output'][0]['content'][0]['text']
    display(result)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
